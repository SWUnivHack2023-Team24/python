```python
import pandas as pd
import requests
```

# OpenWeatherMap

```python
# http://bulk.openweathermap.org/sample/
# 아산, 천안, 청주시 위도/경도 저장
cityList = pd.read_csv('city_list.csv')
```

```python
cheongju = cityList.query('name == ["Asan", "Cheonan", "Cheongju-si"]')
cheongju['lat'] = ""
cheongju['lon'] = ""
```

```python
for i in range(len(cheongju)):
    cheongju['lat'].iloc[i] = eval(cheongju['coord'].iloc[i])['lat']
    cheongju['lon'].iloc[i] = eval(cheongju['coord'].iloc[i])['lon']
```

```python
cheongju
```

```python
w_url = 'http://api.openweathermap.org/data/2.5/air_pollution/forecast?'

# 아산 기준
lat = 'lat=' + str(cheongju['lat'].iloc[0])
lon = '&lon=' + str(cheongju['lon'].iloc[0])
weather_api_key = '&appid=1bd6c3d13fca78e5b8f21cabef64c50c'
```

```python
print(w_url+lat+lon+weather_api_key)
```

```python
result =[]
```

```python
tot = requests.get(w_url+lat+lon+weather_api_key)
data = tot.json()
```

```python
for w in data['list']:
    result.append(
        {
            'dt' : w['dt'],
            'co' : w['components']['co'],
           'no' : w['components']['no'],
           'no2' : w['components']['no2'],
           'o3' : w['components']['o3'],
           'so2' : w['components']['so2'],
           'pm2_5' : w['components']['pm2_5'],
           'pm10' : w['components']['pm10'],
           'nh3' : w['components']['nh3']
        }
    )
```

```python
result_df = pd.DataFrame(result)
```
